## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

## 3. 🔧 Your turn

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

- Redo Discovery A but only for pages with `impressions_90d > 0` — does the correlation change?
- In Discovery B, which `content_type` has the worst CTR *at the same position tier*?
- Find another belief to test: does `content_age_days` relate to `trend_direction`? Does `avg_position` relate to `ctr`?

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
# Your discovery here
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================================
# Load Dataset
# ==========================================================

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("="*70)
print("DATASET OVERVIEW")
print("="*70)
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

# ==========================================================
# Discovery A
# Correlation: Search Volume vs Impressions
# ==========================================================

print("\n" + "="*70)
print("DISCOVERY A")
print("="*70)

corr_all = df["search_volume"].corr(df["impressions_90d"])

filtered = df[df["impressions_90d"] > 0]
corr_filtered = filtered["search_volume"].corr(filtered["impressions_90d"])

print(f"Correlation (All Pages):                {corr_all:.3f}")
print(f"Correlation (Impressions > 0 only):     {corr_filtered:.3f}")

if abs(corr_filtered) < 0.2:
    print("Observation: Search volume is still a very weak predictor of page impressions.")
elif abs(corr_filtered) < 0.5:
    print("Observation: Moderate relationship.")
else:
    print("Observation: Strong relationship.")

# ==========================================================
# Discovery B
# CTR by Position Tier
# ==========================================================

print("\n" + "="*70)
print("DISCOVERY B")
print("="*70)

visible = df[df["impressions_90d"] >= 100]

ctr_by_position = (
    visible.groupby("position_tier")["ctr"]
    .mean()
    .sort_values(ascending=False)
)

print("\nAverage CTR by Position Tier")
print(ctr_by_position.round(4))

plt.figure(figsize=(8,4))
ctr_by_position.plot(kind="bar")
plt.title("Average CTR by Position Tier")
plt.ylabel("CTR")
plt.show()

# ==========================================================
# Worst Content Type at Same Position Tier
# ==========================================================

print("\n" + "="*70)
print("CONTENT TYPE ANALYSIS")
print("="*70)

ctr_content = (
    visible.groupby(["position_tier","content_type"])["ctr"]
    .mean()
    .reset_index()
)

worst = ctr_content.loc[
    ctr_content.groupby("position_tier")["ctr"].idxmin()
]

print(worst)

# ==========================================================
# Discovery C
# Word Count vs Trend Direction
# ==========================================================

print("\n" + "="*70)
print("DISCOVERY C")
print("="*70)

word_count = df.groupby("trend_direction")["word_count"].median()

print(word_count.round(0))

plt.figure(figsize=(7,4))
word_count.plot(kind="bar")
plt.title("Median Word Count by Trend Direction")
plt.ylabel("Median Word Count")
plt.show()

# ==========================================================
# Content Age vs Trend Direction
# ==========================================================

print("\n" + "="*70)
print("CONTENT AGE ANALYSIS")
print("="*70)

content_age = df.groupby("trend_direction")["content_age_days"].median()

print(content_age.round(0))

plt.figure(figsize=(7,4))
content_age.plot(kind="bar")
plt.title("Median Content Age by Trend Direction")
plt.ylabel("Days")
plt.show()

# ==========================================================
# Average Position vs CTR
# ==========================================================

print("\n" + "="*70)
print("AVERAGE POSITION VS CTR")
print("="*70)

position_corr = df["avg_position"].corr(df["ctr"])

print(f"Correlation between Average Position and CTR: {position_corr:.3f}")

plt.figure(figsize=(6,5))
plt.scatter(df["avg_position"], df["ctr"], alpha=0.3)
plt.xlabel("Average Position")
plt.ylabel("CTR")
plt.title("Average Position vs CTR")
plt.show()

# ==========================================================
# Summary
# ==========================================================

print("\n" + "="*70)
print("SUMMARY OF FINDINGS")
print("="*70)

print("""
1. Search volume has little correlation with actual page impressions,
   even after filtering pages with impressions greater than zero.

2. CTR decreases as search position becomes worse, confirming the
   expected CTR cliff.

3. Different content types perform differently even at the same
   position tier. Some content types consistently receive lower CTR.

4. Declining and growing pages have similar median word counts,
   suggesting that simply writing longer content is unlikely to
   improve performance.

5. Content age varies across trend directions and may influence
   whether content grows or declines.

6. Average search position is strongly related to CTR; pages ranking
   higher generally receive substantially more clicks.
""")
```
